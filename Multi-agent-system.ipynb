```python
#Task 1 -- the first task is of data compilation, where given the price history of iphones and the time-line of major world events, get a combined data
data_insights_task_1 = Task(
    description=(
        "You have been provided with two resources, a dataset of price changes in different iphone models and makes as observed over the past few years, and a list of all major events happening across the Apple Company, and the World"
        "Your task to look at both the data, match their timelines and then try to infern if the change in prices of the models are due to the event list provided to you."
        "Since the events that happen at Apple are periodic over time (generally about a year) compare how the changes are carried over from one year to the other."
        "Try to smooth out the data with the help of justification from major world events, since our main focus is periodic apple events."
        "Irrespective of the changes, try to infern why did it or did it not happen exactly the same as previous year."
        "You have to make an indepth-analysis about the relation between the affects of periodic events and the prices change of the models"
    ),
    expected_output=(
        "A comprehensive report over the relation between the fluctuation of the prices and periodic events."
        "This should also include your insights as to what you think can and would happen in the future, based on the obervations"
        "Highlight potential major factors that you belive were the leading causes."
    ),
    tools=[directory_read_tool_1, file_read_tool],
    agent=data_compilation_agent_1,
    async_execution=True,
    output_file="./Reports/Iphone1_report.md"
)
```
